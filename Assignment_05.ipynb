**Since there are too many variables to explore, many of which are not interesting. Create a smaller data frame with the following 9 variables:** *UNITID, INSTNM, CITY, STABBR, ZIP, ADM_RATE, UGDS, TUITIONFEE_IN, MN_EARN_WNE_P10* 

```
df_2 = df_1[["UNITID", "INSTNM", "CITY", "STABBR", "ZIP", "ADM_RATE", "UGDS", "TUITIONFEE_IN", "MN_EARN_WNE_P10" ]]
print(df_2)
df_3 = pd.DataFrame(df_2)
```

**Defination of the 9 column headings**

*   UNITID : Unit ID for the institution (School)
*   INSTNM : Name of Institution
*   CITY : City where the Institution is located
*   STABBR : Abbrivation of the name of the state where the institution is located. 
*   ZIP : Zip code of where the institution is located.
*   ADM_RATE : Admission rate for each campus which is number of admitted undergraduate students divided by the number of undergraduate application. 
*   UGDS : This is the number of degree or certificate seeking undergraduates enrolled. 
*   TUITIONFEE_IN : This is the tuition and fees for instate students. 
*   MN_EARN_WNE_P10 : This represents the mean institutionally aggrigated earnings for all students that had recieved federal aid previously and are currently employed and not enrolled in school now.

**Find and display UMBC**

```
Select_UMBC = df_2.loc[df_2['INSTNM'] =='University of Maryland-Baltimore County']
pd.option_context('display.max_columns', None)
#pd.set_option('display.width', None)
print(Select_UMBC)
```

**Find Univ. of Maryland College Park**

**Find and display Loyola College Maryland**

**Find and display Johns Hopkins**

```
Select_JHU = df_2.loc[df_2['INSTNM'] == 'Johns Hopkins University']
print(Select_JHU)
```

**Calculate Min, Max and Average for the in-state tuitions of the 4 institutions.**

```
frames = [Select_JHU, Select_Loyola, Select_UMBC, Select_UMD]
result = pd.concat(frames)
print(result)
result.shape

Ave_tui = result['TUITIONFEE_IN'].mean()
Min_tui = result['TUITIONFEE_IN'].min()
Max_tui = result['TUITIONFEE_IN'].max()
print('Average tuition is: ', Ave_tui)
print('Minimum Tuition is: ', Min_tui)
print('Maximum Tuition is: ', Max_tui)
```
