# Import the required libraries. 

```python
import pandas as pd
file_path = '/content/drive/MyDrive/Colab Notebooks/census_cost.txt'
```

# Read the text file skipping the first row.

```python
df = pd.read_csv(file_path, delimiter='\t', skiprows =1)
df
```

# Use the "replace" function in python to replace ',' and '*' with nothing. 

```python
df['Total Population']= df['Total Population'].str.replace(',','')
df['Census Year']= df['Census Year'].str.replace('*','')

df
```

# Define a function to clean the cost column.

```python
def cleanse_cost(raw_cost):
    cost = raw_cost.lstrip("$").replace(",","")
    if  raw_cost.find("4.5") != -1:          # if "4.5" in raw_cost:
        cost = "45" + "0" * 8
    elif raw_cost.find("Billion") != -1:     # elif "Billion" in raw_cost: 
        cost = "13" + "0" * 9
    return cost
```

# Run the previously created function in the dataframe using "apply"

```python
df['Census Cost']=df['Census Cost'].apply(cleanse_cost)
df
```

# Create a function to clean up the Average Cost column and run the function using "apply"

```python
def cleanse_average(average_cost):
  ave_cost = average_cost.replace('$','')
  if(-1 != ave_cost.find('cent')):
    ave_cost= float(ave_cost.split(' ')[0])/100

  return ave_cost
```

```python
df['Average Cost Per Person']=df['Average Cost Per Person'].apply(cleanse_average)
df
```
